```python
dados_reservas = []
for i in range(1,400):
    url = 'https://booking.onhousi.com/reservations?token=5b1f1f4bf68208cfe93445c034072df0c5e50002&limit=100&page={}'.format(i)
    res = requests.get(url)
    lista_reservas = json.loads(res.text)['items']
    for x in range(0,100):
        uni = {}
        try:
            uni['status'] = lista_reservas[x]['status']
        except:
            pass 
        try:
            uni['channel'] = lista_reservas[x]['channel']
        except:
            pass 
        try:
            uni['isSubscription'] = lista_reservas[x]['isSubscription']
        except:
            pass 
        try:
            uni['reserva'] = lista_reservas[x]['aliasId']
        except:
            pass 
        try:
            uni['aliasId'] = lista_reservas[x]['property']['aliasId']
        except:
            pass 
        try:
            uni['name'] = lista_reservas[x]['property']['name']
        except:
            pass
        try:
            uni['_id_property'] = lista_reservas[x]['property']['_id']
        except:
            pass
        try:
            uni['cidade'] = lista_reservas[x]['property']['address']['city']['name']
        except:
            pass 
        try:
            uni['bairro'] = lista_reservas[x]['property']['address']['neighborhood']['name']
        except:
            pass 
        try:
            uni['rua'] = lista_reservas[x]['property']['address']['street']
        except:
            pass 
        try:
            uni['CEP'] = lista_reservas[x]['property']['address']['zipCode']
        except:
            pass 
        try:
            uni['Id_parceiro'] = lista_reservas[x]['partner']['aliasId']
        except:
            pass 
        try:
            uni['parceiro1'] = lista_reservas[x]['partner']['name1']
        except:
            pass 
        try:
            uni['parceiro2'] = lista_reservas[x]['partner']['name2']
        except:
            pass
        try:
            uni['cupom_name'] = lista_reservas[x]['coupon']['name']
        except:
            pass
        try:
            uni['cupom_tipo'] = lista_reservas[x]['coupon']['kind']
        except:
            pass
        try:
            uni['cupom_amount'] = lista_reservas[x]['coupon']['amount']
        except:
            pass
        try:
            uni['criacao'] = lista_reservas[x]['createdAt']
        except:
            pass
        try:
            uni['entrada'] = lista_reservas[x]['checkIn']['date']
        except:
            pass
        try:
            uni['saida'] = lista_reservas[x]['checkOut']['date']
        except:
            pass
        try:
            uni['guest_date'] = lista_reservas[x]['guests']['date']
        except:
            pass
        
        try:
            uni['_id_guest'] = lista_reservas[x]['user']['_id']
        except:
            pass
        try:
            uni['guest_name'] = lista_reservas[x]['user']['firstName']
        except:
            pass
        try:
            uni['guest_surname'] = lista_reservas[x]['user']['lastName']
        except:
            pass
        try:
            uni['guest_email'] = lista_reservas[x]['user']['email']
        except:
            pass
        try:
            uni['cost_total'] = lista_reservas[x]['cost']['total']['amount']
        except:
            pass
        try:
            uni['cost_gross'] = lista_reservas[x]['cost']['grossAmount']['amount']
        except:
            pass
        try:
            uni['cost_net'] = lista_reservas[x]['cost']['netAmount']['amount']
        except:
            pass
        try:
            uni['fee'] = lista_reservas[x]['cost']['fees']['amount']['amount']
        except:
            pass
        try:
            uni['sellerPerson'] = lista_reservas[x]['sellerPerson']['name']
        except:
            pass
        dados_reservas.append(uni)
        
```

```python
df_reservas = pd.DataFrame()
df_reservas = df_reservas.append(dados_reservas)
```
